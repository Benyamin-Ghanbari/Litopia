# CPU-GPU extraction func.

```python
def extract_info(cpu_name):
   
    brand, model, speed = None, None, None

    # Intel CPUs
    match_core = re.match(r"(?i)Intel\s+Core\s+([^\s]+)\s*([^\s]+)?\s*(\d+(?:\.\d+)?)GHz", cpu_name)
    if match_core:
        brand = "Intel"
        model = f"Core {match_core.group(1)}"
        speed = match_core.group(3)
        if match_core.group(2):
            model += " " + match_core.group(2)
        return brand, model, speed

    match_celeron = re.match("(?i)Intel\s+Celeron\s+(Dual|Quad) Core\s+([^\s]+)\s*(\d+(?:\.\d+)?)GHz", cpu_name)
    if match_celeron:
        brand = "Intel"
        model = f"Celeron {match_celeron.group(1)} Core {match_celeron.group(2)}"
        speed = match_celeron.group(3)
        return brand, model, speed

    match_pentium = re.match(r"(?i)Intel\s+Pentium\s+(Dual|Quad) Core\s+([^\s]+)\s*(\d+(?:\.\d+)?)GHz", cpu_name)
    if match_pentium:
        brand = "Intel"
        model = f"Pentium {match_pentium.group(1)} Core {match_pentium.group(2)}"
        speed = match_pentium.group(3)
        return brand, model, speed

    match_atom = re.match(r"(?i)Intel\s+Atom\s+([^\s]+)\s*(\d+(?:\.\d+)?)GHz", cpu_name)
    if match_atom:
        brand = "Intel"
        model = f"Atom {match_atom.group(1)}"
        speed = match_atom.group(2)
        return brand, model, speed

    match_xeon = re.match(r"(?i)Intel\s+Xeon\s+([^\s]+)\s*([^\s]+)?\s*(\d+(?:\.\d+)?)GHz", cpu_name)
    if match_xeon:
        brand = "Intel"
        model = f"Xeon {match_xeon.group(1)}"
        speed = match_xeon.group(3)
        if match_xeon.group(2):
            model += " " + match_xeon.group(2)
        return brand, model, speed

    # AMD CPUs
    match_amd = re.match(r"(?i)AMD\s+([^\s]+)\s*([^\s]+)?\s*(\d+(?:\.\d+)?)GHz", cpu_name)
    if match_amd:
        brand = "AMD"
        model = match_amd.group(1)
        speed = match_amd.group(3)
        if match_amd.group(2):
            model += " " + match_amd.group(2)
        return brand, model, speed

    # Samsung products
    match_samsung = re.match(r"(?i)Samsung\s+([^\s]+)\s*([^\s]+)?\s*(\d+(?:\.\d+)?)GHz", cpu_name)
    if match_samsung:
        brand = "Samsung"
        model = match_samsung.group(1)
        speed = match_samsung.group(3)
        if match_samsung.group(2):
            model += " " + match_samsung.group(2)
        return brand, model, speed

    return brand, model, speed
```

```python
data[['CPU_Brand', 'CPU_Model', 'CPU_Speed']] = data['CPU'].apply(lambda x: pd.Series(extract_info(x)))
```

#  correcting storage

```python
ssd_lst = list()
hdd_lst = list()
hybrid_lst = list()
for i in range(data.shape[0]):
    storage = data.Storage[i].split(sep = '+')
    if len(storage) == 1:
        if 'SSD' in storage[0]:
            ssd_lst.append(storage[0].split(sep = ' ')[0])
            hdd_lst.append(0)
            hybrid_lst.append(0)
        elif 'HDD' in storage[0]:
            hdd_lst.append(storage[0].split(sep = ' ')[0])
            ssd_lst.append(0)
            hybrid_lst.append(0)
        elif 'Flash' in storage[0]:
            ssd_lst.append(storage[0].split(sep = ' ')[0])
            hdd_lst.append(0)
            hybrid_lst.append(0)
        elif 'Hybrid' in storage[0]:
            hybrid_lst.append(storage[0].split(sep = ' ')[0])
            ssd_lst.append(0)
            hdd_lst.append(0)
    else:
        stor1 = storage[0]
        stor2 = storage[1]
        st1 = storage[0].split(sep = ' ')[0]
        st2 = (storage[1].strip()).split(sep = ' ')[0]
        if 'SSD' in stor1:
            ssd_lst.append(st1)
            if 'SSD' in stor2:
                ssd_lst[i] += ' '
                ssd_lst[i] += st2
        elif 'SSD' in stor2:
            ssd_lst.append(st2)   
        elif 'Flash' in stor1:
            ssd_lst.append(st1)
        elif 'Flash' in stor2:
            ssd_lst.append(st2)
        else:
            ssd_lst.append(0)
        
  
        if 'HDD' in stor1:
            hdd_lst.append(st1)
            if 'HDD' in stor2:
                hdd_lst[i] += ' '
                hdd_lst[i] += st2
        elif 'HDD' in stor2:
            hdd_lst.append(st2)
        else:
            hdd_lst.append(0)
            
        if 'Hybrid' in stor1:
            hybrid_lst.append(st1)
        elif 'Hybrid' in stor2:
            hybrid_lst.append(st2)
        else:
            hybrid_lst.append(0)
            
```

```python
storage_df = pd.DataFrame({'storage': data.Storage,
                          'SSD': ssd_lst,
                          'HDD': hdd_lst,
                          'Hybrid': hybrid_lst})
```

```python
def convert(st):
    if 'TB' in st:
        a = int(st.replace('TB',''))*1000
    if 'GB' in st:
        a = int(st.replace('GB',''))
    return a
```

```python
def storage_int(stor):
    if stor == 0:
        re = 0
    else:
        stor_l = stor.split(sep = ' ')
        if len(stor_l) == 2:
            st1 = convert(stor_l[0])
            st2 = convert(stor_l[1])
            re = st1 + st2
        else:
            st = stor
            re = convert(st)
    return re
```

```python
def hybrid(hyb):
    s = int((1/4)*(convert(hyb)))
    h = convert(hyb) - s
    return [s, h]
```

```python
SSD_lst = list()
HDD_lst = list()
for i in range(storage_df.shape[0]):
    ssd = storage_df.SSD[i]
    hdd = storage_df.HDD[i]
    hyb = storage_df.Hybrid[i]
    SSD_lst.append(storage_int(ssd))
    HDD_lst.append(storage_int(hdd))
    if hyb != 0:
        lst = hybrid(hyb)
        SSD_lst[i] += lst[0]
        HDD_lst[i] += lst[1]
        
```

```python
storage_df['new ssd'] = SSD_lst
storage_df['new hdd'] = HDD_lst
```
