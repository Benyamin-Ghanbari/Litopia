```python
screentypelist = counts.index.to_list()
import re

def group_screen_type(screen_type):
    if re.search(r'IPS|LED-backlit|WLED|BrightView|Liquid Retina|OLED|Truelife|LCD|PixelSense|AMOLED|TruBrite|Retina|Gorilla Glass|InfinityEdge', screen_type):
        return 'IPS or LED-backlit'
    elif re.search(r'TFT|TN', screen_type):
        return 'TFT or TN'
    elif re.search(r'HD|Full HD|FHD', screen_type):
        return 'HD or Full HD'
    elif re.search(r'WV|SVA', screen_type):
        return 'WV or SVA'
    elif re.search(r'Anti-Glare|True Life', screen_type):
        return 'Anti-Glare or True Life'
    elif re.search(r'WLED|BrightView', screen_type):
        return 'WLED or BrightView'
    elif re.search(r'4K|UHD', screen_type):
        return '4K or UHD'
    elif re.search(r'QHD|WUXGA', screen_type):
        return 'QHD or WUXGA'
    elif re.search(r'Ultra|TruBrite', screen_type):
        return 'Ultra or TruBrite'
    elif re.search(r'XDR', screen_type):
        return 'XDR'
    else:
        return 'Others'

# Group the screen types
grouped_screen_types = [group_screen_type(screen_type) for screen_type in screentypelist]

# Find unique grouped screen types
unique_grouped_screen_types = list(set(grouped_screen_types))

print(unique_grouped_screen_types)
```

```python
# Create a dictionary mapping the original screen types to the grouped ones
screen_type_dict = {screen_type: group_screen_type(screen_type) for screen_type in screentypelist}

# Replace the original screen types with the grouped ones in the DataFrame
df['Screen_type'] = df['Screen_type'].replace(screen_type_dict)
```

## Storage type 

```python
def clean_column(df, column):
    df[column] = df[column].replace({
        'SSDHDD': 'SSD, HDD',
        'SSDSATA': 'SSD, SATA',
        'SSDحافظه‌های هیبردی': 'SSD, Hybrid Storage',
        'SSDHDDحافظه‌های هیبردی': 'SSD, HDD, Hybrid Storage',
        'SSDSSHD': 'SSD, SSHD',
        'SSHDSATA': 'SSHD, SATA',
        'SSDSATAحافظه‌های هیبردی': 'SSD, SATA, Hybrid Storage',
        'حافظه‌های هیبردی':'Hybrid'
    })
    return df

df = clean_column(df, 'Storage_type')
df['Storage_type']=df['Storage_type'].replace(['SSD, Hybrid Storage','SSD, HDD, Hybrid Storage',
                                              'SSHD','SSD, SSHD'],'Hybrid')
df['Storage_type']=df['Storage_type'].replace('','eMMC')
df['Storage_type']=df['Storage_type'].replace(['SSD, HDD','Hybrid'],'Hybrid')
df['Storage_type']=df['Storage_type'].replace(['SSD, SATA, Hybrid Storage','SSHD, SATA'],'SSD, SATA')
```

```python
import pandas as pd

def replace_nan(df):
    df.loc[(df['OS'].isnull()) & (df['CPU_Brand'] == 'Apple'), 'OS'] = 'Apple Mac OS'
    df.loc[(df['OS'].isnull()) & (df['CPU_Brand'] == 'INTEL'), 'OS'] = 'Windows X'
    df.loc[(df['OS'].isnull()) & (df['CPU_Brand'] == 'AMD'), 'OS'] = 'Windows Y'
    return df

df = replace_nan(df)
```
