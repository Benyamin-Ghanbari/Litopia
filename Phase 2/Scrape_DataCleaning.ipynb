# Price Cleaning

```python
df['original_price'] = df['original_price'].astype(str)
df['after_discount'] = df['after_discount'].astype(str)

df['original_price'] = df['original_price'].str.replace(',', '')
df['after_discount'] = df['after_discount'].str.replace(',', '')
df['original_price'] = df['original_price'].str.replace('تومان', '')
df['after_discount'] = df['after_discount'].str.replace('تومان', '')
df['original_price'] = df['original_price'].str.replace(r'\D', '')
df['after_discount'] = df['after_discount'].str.replace(r'\D', '')

df['original_price'] = pd.to_numeric(df['original_price'], errors='coerce')
df['after_discount'] = pd.to_numeric(df['after_discount'], errors='coerce')

df['original_price'] = df['original_price'].apply(lambda x: int(x) if pd.notnull(x) else x)

# Convert 'after_discount' to int, ignoring null values
df['after_discount'] = df['after_discount'].apply(lambda x: int(x) if pd.notnull(x) else x)
```

# Weight Cleaning

```python
df['Weight'] = df['Weight'].astype(str)
df['Weight'] = df['Weight'].str.replace('کیلوگرم','KG')
df['Weight'] = df['Weight'].str.replace('کلیوگرم','KG')
df['Weight'] = df['Weight'].str.replace('کیلو گرم', 'KG')
df['Weight'] = df['Weight'].str.replace('گ', 'G')
```

```python
import re
def check_and_clean_Weight(df, column):
    pattern = r'^\d+\.\d+\sKG$'
    df = df[df[column].apply(lambda x: bool(re.match(pattern, str(x))))]
    
    return df
df=check_and_clean_Weight(df,'Weight')
```

# Screen size Cleaning

```python
def clean_screen_size(df, column):
    df[column] = df[column].str.replace(' اینچ', '').astype(float)
    return df

df = clean_screen_size(df, 'Screen_size')
df.head()
```

```python
def clean_ram(df, column):
    df[column] = df[column].str.replace(' گیگابایت', '').astype(float)
    return df

df = clean_ram(df, 'RAM')
df.head()
```

# Storage cleaning

```python
def clean_Storage(df, column):
    df[column] = df[column].str.replace(' گیگابایت', '')
    df[column] = df[column].str.replace(' ترابایت', '').astype(float) * 1024
    return df

df = clean_Storage(df, 'Storage_amount')
df.head()
```

# Discount calculator

```python
def discount_percentage(original, after):
    if pd.notnull(original) and pd.notnull(after):
        return round((original - after) / original * 100, 2)
    # Otherwise, return 0
    else:
        return 0

# Apply the function to the original_price and after_discount columns and create a new column called discount_percent
df['discount_percent'] = df.apply(lambda x: discount_percentage(x['original_price'], x['after_discount']), axis=1)
```
