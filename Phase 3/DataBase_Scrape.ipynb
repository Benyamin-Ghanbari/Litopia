```python
def insert(sql, df):
    cols = list(df.columns)
    rec = list()
    for i in range(df.shape[0]):
        rec = list()
        for c in cols:
            col_type = str(df[c].dtype)
            if col_type == 'int64':
                r = int(df[c][i])
            elif col_type == 'float64':
                r = float(df[c][i])
            else:
                r = df[c][i]
            rec.append(r)
        record = tuple(rec)
        mycursor.execute(sql, record)
    mydb.commit()  
```
